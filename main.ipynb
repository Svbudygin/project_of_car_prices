# Data cleanup and Transformation
## In this table, there are numerous missing values. Let me work with them
1) Initially, if the year information is absent, it would be great to compute the mean year for cars with the same name (model). However, if a vehicle's name uniquely identified, then finding its year becomes impossible, so I decided to drop this rows from my dataset.
2) Similarly, I will apply the same strategy to the power column. To execute this, I will construct a pivot table and introduce two new columns, that will display the average horsepower and year for each car name (model).
3) As for mileage, a NaN value implies a new vehicle with the mileage value of 0.
4) Non-numeric columns will have NaN entries replaced with 'information not given,' as these columns serve just additional details.
5) In the interest of making analysis more simple, I have dropped columns such as 'description,' 'link,' 'location,' and 'parse_date'.
6) Furthermore, I have standardized the column names to lowercase.
7) For "fuel type", "transmission", "name", "brand" and "price", I removed rows containing deleted NaN entries. This step is very important because attempting to replace NaN values in these columns is not possible due to lack of information.
8) I have also change NaN values in the 'color' column by "Белый," following the common practice on Auto.ru, where an absence of color denotes as a white vehicle.
9) Additionally, I have reset the indices and reform certain columns, including 'power,' 'year,' and 'mileage,' to integer format.

```python
def int_mean(arg):
    try:
        return int(arg.mean())
    except ValueError:
        return np.nan


pivot_data = data.pivot_table(index='name', values=['year', 'power', 'price'], aggfunc=int_mean)
print(pivot_data)
```

```python
data = data.merge(pivot_data, on='name', how='left', suffixes=('', '_average'))
data.loc[data['year'].isna(), 'year'] = data.year_average
data.loc[data['power'].isna(), 'power'] = data.power_average
data.head()
```

```python
print(data.shape)
data.dropna(axis=0, inplace=True, how='any', subset=['year', 'power'])
print(data.shape)
```

```python
NG = 'information not given'
data = data.fillna(
    {'mileage': 0, 'bodyType': NG, 'vehicleConfiguration': NG, 'engineName': NG, 'engineDisplacement': NG,
     'color': 'Белый'})
data.head()
```

```python
data = data.drop(['description', 'link', 'location', 'parse_date', 'vehicleConfiguration', "engineName", "engineDisplacement", 'date'], axis=1)
data.head()
```

```python
data.columns = data.columns.str.lower()
data.dropna(axis=0, inplace=True, how='any', subset=['fueltype', 'transmission', 'name', 'brand', 'price'])
data = data.reset_index(drop=True)
data.year = data.year.apply(int)
data.power = data.power.apply(int)
data.mileage = data.mileage.apply(int)
data.tail()
```

```python
qx = data['brand'].unique()[7:50]
data = data.loc[data.brand.isin(set(qx) | {'Ferrari', "Lamborghini"})]
data.head()
```

# My Hypothesis
Regarding this hypothesis, I believe that the price of a car depends on several factors:

When it comes to vehicle pricing, brand luxury stands for 10%, with fuel type and transmission sharing a similar percentage, taking another 10%. Power plays a great role, accounting for 20%, and more significantly 30% is related to mileage. The remaining 30% is distributed among various other parameters.

My idea involves restructuring our data set. I plan to introduce a new columns that contains types of fuel, represented digitally. Additionally, I aim to rearrange car brands into three types based on their luxury status, taking into account car prices and horsepower. To uncover the truth I intend to  explore the correlations.


```python
def change_fuel(fuel):
    if fuel.lower() == "бензин":
        return 1
    if fuel.lower() == 'дизель':
        return 2
    return 3


def change_tr(tr):
    if tr == 'Механика':
        return 1
    return 2


data['trms'] = data.transmission.apply(change_tr)
data['fuel'] = data.fueltype.apply(change_fuel)
data.head()
```

```python
def int_mean(arg):
    return int(arg.mean())


def lux(car_bruum):
    global cm, pm
    return car_bruum / cm * 2


def lux1(pow):
    global cm, pm
    return pow / pm


pivot_df = data.pivot_table(index='brand', values=['price', 'power', ], aggfunc=int_mean)
data = data.merge(pivot_df, on='brand', how='left', suffixes=('', '_avrg_brand'))
cm = data.price_avrg_brand.mean()
pm = data.power_avrg_brand.mean()

data.tail()
```

```python
data['luxury'] = data['price_avrg_brand'].apply(lux) + data['power_avrg_brand'].apply(lux1)
data.head()
```

```python
pd.plotting.scatter_matrix(pd.DataFrame(data=data, columns=['price', 'luxury', 'fuel', 'trms', 'power', 'mileage']), figsize=(20, 10))
plt.savefig('11.png')
plt.show()
```

Returning to my hypothesis, let's analyse its key aspects:


1) The dependence of price on fuel and transmission close to each other, sharing a proportional percentage.
2) Interestingly, this percentage weighs almost like the ratio given to the luxury factor in the pricing dependance equation.
3) On the other hand, the power of a given vehicle is expected to have twice the impact on its price compared to the combined effects of fuel and transmission.
4) Significantly, and according to my assumption, the main factor affecting price is the mileage, which waits three times more than luxury.

Let's delve into correlations to find the truth.

```python
plt.figure()
sns.heatmap(data[['price', 'luxury', 'fuel', 'trms', 'power', 'mileage']].corr(), annot=True, fmt='.4f', linewidths=.5, cmap='winter')
plt.savefig('12.png')
plt.show()
```

In my opinion, only correlations with 0,5 and bigger can be caused by some real connections, but other correlations are just coincidences. 	
* Price vs. Luxury (0.509580):
There is a positive correlation between price and luxury. This suggests that as the price of a car increases, it is likely to be associated with more luxury cars.
* 	Price vs. Fuel Efficiency (0.253799):
There is a positive correlation between price and fuel efficiency, although it is relatively weak. This could mean that higher-priced cars tend to have slightly better fuel efficiency, but the relationship is not very strong.
* 	Price vs. Transmission (0.289304):
There is a positive correlation between price and transmission. This could show that higher-priced cars are more likely to have automatic transmission.
* 	Price vs. Power (0.608761):
There is a strong positive correlation between price and power. This suggests that more expensive cars have more engine power.
* 	Price vs. Mileage (-0.267206):
There is a weak negative correlation between price and mileage. This implies that as the price of a car increases, it is likely to be associated with lower mileage.

The most significant correlations with the price has power (0.6), followed by luxury (0.5), and then transmission (0.28). Surprisingly, mileage (-0.26) and the type of fuel (0.25) exhibit comparatively weaker relationships with price...

Contrary to my expectations, the analysis proves that power plays a more crucial role than mileage in influencing the price. Furthermore, it is interesting to note that price doesn't significantly depend on mileage. Transmission and fuel show similar correlation grades, as we expected. A new discovery for me is the significant influence luxury car brands have on the pricing process.
